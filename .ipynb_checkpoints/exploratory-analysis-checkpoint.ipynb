```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../Data/data.csv")
data.head()
```

## Which characters appeared on the most episodes?

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
a = pysqldf("SELECT character, gender, count(distinct season_episode) as total_appearances FROM data GROUP BY 1,2 ORDER BY 3 DESC LIMIT 10;")
a
```

Top 5: 2 male, 3 female
<br>
Top 10: 6 male, 4 female

```python
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
%matplotlib inline
```

```python
sns.barplot(x="total_appearances", y="character", hue="gender", data=a)
```

## Which characters spoke in terms of most words spoken?

```python
b = pysqldf("SELECT character, gender, sum(n_words) as total_words_spoken FROM data GROUP BY 1,2 ORDER BY 3 DESC LIMIT 10;")
b
```

Top 5: 3 male, 2 female
<br>
Top 10: 6 male, 4 female

```python
sns.barplot(x="total_words_spoken", y="character", hue="gender", data=b)
```

## Which characters had the largest vocabulary

```python
character_lines = data.groupby(['character','gender'])['line'].apply(lambda x: ' '.join(x)).reset_index()
character_lines.head()
```

```python
character_lines['nunique_words']  = character_lines['line'].apply(lambda x: len(set(x.split())))
character_lines.sort_values(by='nunique_words', ascending=False).head() # looks good
```

```python
c = pysqldf("SELECT character, gender, nunique_words FROM character_lines ORDER BY 3 DESC LIMIT 10;")
c
```

Top 5: 4 male, 1 female
<br>
Top 10: 7 male, 3 female

```python
sns.barplot(x="nunique_words", y="character", hue="gender", data=c)
```
